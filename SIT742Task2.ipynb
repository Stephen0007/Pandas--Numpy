# Part 2. Text Analysis

## 2.1 extract high requency words in description

```python
import re
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.probability import *
from itertools import chain
#from tqdm import tqdm
import codecs

```

```python
with open('stopwords.txt') as f:
    stop_words = f.read().splitlines()
stop_words = set(stop_words)

#By using the open and read function, read the stop word file 
```

```python
# write your code here
# define your tokenize

tokenizer = RegexpTokenizer(r"\w+(?:[-']\w+)?")
Sentan_List = []
for title in d11['description']:
    Sentan=[w for w in tokenizer.tokenize(title) if w.lower() not in stop_words]
    Sentan_List.append(Sentan)
    #tokenisedword = [word for word in Sentan for sentan in Sentan_List]
tokenisedword =list(chain.from_iterable(Sentan_List))

#By using the Regular expression tokenizer to form a search pattern
```

```python
#find top common words with document frequencies > 5000
# you may use function FreqDist() and sort()

from itertools import takewhile
fdd = FreqDist(tokenisedword)
dat1 = fdd.most_common()
top_common = dict(takewhile(lambda x : x[1] > 5000, dat1))
top_common
key1 = top_common.keys()


sorted_top_common = {}

for key in sorted(key1):
    sorted_top_common[key] = top_common[key]
    
    
sorted_top_common

#By using the FreqDist we need to pull down the keywords with the frequency > 5000 and dump them in top_common
```

```python
# save your table to 'top_common_words.
with open('Highfreq.txt', 'w') as file:
    file.write(json.dumps(sorted_top_common))
    
#Store the top_common words in the text file named 'top_common_words'
```

## 2.2 Find key words for describing Shiraz using TF-IDF

```python
# select 'description' from 'variety' eqaul to  'Shiraz'
shiraz_sentances = d11[d11['variety'] == 'Shiraz']['description']
shiraz_sentances

#Find the keyword decscribtion from the data frame column named 'variety'
```

```python
# use TfidfVectorizer to calculate TF-IDF score
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(analyzer='word', stop_words = 'english')
Y = tfidf.fit_transform(shiraz_sentances)
idf = tfidf.idf_
tf1_value = dict(zip(tfidf.get_feature_names(),idf))
tf1_value.items

#Calculate the Tfidf score
```

```python
# find words with TF-IDF score >0.4 and sort them
#tf1.value.items()
word_in_four = list((key,value.round(2)) for (key,value) in tf1_value.items() if value>0.4)
word_in_four.sort()
word_in_four

#List down the key values > 0.4
```

```python
# save your table to 'key_Shiraz.txt'   
with open('Shirazkey.txt', 'w') as file:
    file.write(json.dumps(word_in_four))
```
